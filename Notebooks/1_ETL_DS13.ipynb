#  **EDA - Unnesting Datasets 1 and 3**

We have two other .json files that need to be explored and unnested:

- steam_games.json
- users_items.json

### 1. Data Load:

### *1.1. Steam_games Dataset:*

```python
import pandas as pd
import json

data = []
with open('/content/drive/MyDrive/Colab Notebooks/Dataset/steam_games.json', 'r') as file:
    for line in file:
        data.append(json.loads(line))

df_steam_games = pd.DataFrame(data)
df_steam_games
```

```python
df_steam_games.info()
```

### *1.2. Users_items Dataset:*

```python
# Due to the file users_items.json has some dictionaries and list nested into a table that is not a .json format,
# we may need to use the library "ast". The ast.literal_eval() function is capable of evaluating Python literal
# expressions, including dictionaries and lists. By reading each line of the file and applying ast.literal_eval(),
# you are interpreting each line as a Python literal expression and converting it into a Python object.

import ast

data = []
with open('/content/drive/MyDrive/Colab Notebooks/Dataset/users_items.json', 'r', encoding='utf-8') as file:
    for line in file.readlines():
        data.append(ast.literal_eval(line))

df_users_items = pd.DataFrame(data)
df_users_items
```

```python
df_users_items.info()
```

### 2. Data Transformation:

### *2.1. Steam_games Dataset:*

We can see that in the df_steam_games we have 120,445 entries, but the first rows are null in every column. So it would be a good idea to eliminate them:

```python
df_steam_games1 = df_steam_games.copy()
df_steam_games1
```

```python
# Mask creation

mask1 = df_steam_games['publisher'].isna() & df_steam_games['genres'].isna() & df_steam_games['app_name'].isna() & df_steam_games['title'].isna() & df_steam_games['url'].isna() & df_steam_games['release_date'].isna() & df_steam_games['tags'].isna() & df_steam_games['reviews_url'].isna() & df_steam_games['specs'].isna() & df_steam_games['price'].isna() & df_steam_games['early_access'].isna() & df_steam_games['id'].isna() & df_steam_games['developer'].isna()
df_steam_games[mask1]
```

We can see that those entries that have all its columns with NaN are the first 88310 (from row 0 to row 88309). Let's eliminate them from te dataframe.

```python
df_steam_games1.drop(df_steam_games1.index[0:88310], inplace=True)
df_steam_games1
```

We have now 32135 entries. Now let's reset the index of this dataframe.

```python
df_steam_games1.reset_index(drop=True, inplace=True)
df_steam_games1
```

Now, that the dataframe is organized cleaned from the the entries without information, we are going to proceed to load it into a CSV file as follows:

```python
df_steam_games1.to_csv('steam_games_final.csv', index=False)
```

### *2.2. Users_items Dataset:*

```python
df_users_items1 = df_users_items.copy()
df_users_items1
```

We can see that the column "items" has a list of dictionaries per userid. Let's unnest this column.

```python
# We expand the dataframe users_items in the column "items" in order to convert each element of the lists into one unique row.

df_users_items_Mod_exploded = df_users_items1.explode('items')
df_users_items_Mod_exploded
```

```python
# Now we need to convert all the dictionaries of the column "items" through the function json.normalized

df_users_items_Mod_normalized = pd.json_normalize(df_users_items_Mod_exploded['items'])
df_users_items_Mod_normalized
```

```python
#Reset the index for each dataframe:

df_users_items_Mod_exploded.reset_index(drop=True, inplace=True)
df_users_items_Mod_normalized.reset_index(drop=True, inplace=True)
```

```python
#We concatenate both dataframes:
df_users_items_expanded = pd.concat([df_users_items_Mod_exploded.drop('items',axis=1),df_users_items_Mod_normalized], axis=1)
df_users_items_expanded
```

```python
df_users_items_expanded.to_csv('users_items_final.csv', index=False)
```
