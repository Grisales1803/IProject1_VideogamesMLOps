#  **Exploratory Data Analysis (EDA)**

Exploratory Data Analysis (EDA) is crucial in data science and analytics for several important reasons:

- **Data Understanding:** EDA helps us gain a deep understanding of our dataset. We can discover the structure, patterns, and relationships within the data. This understanding is essential before we can proceed with more advanced analysis or modeling.

- **Data Quality Check:** EDA allows us to identify data quality issues such as missing values, outliers, and inconsistencies. Addressing these issues early in the analysis process is critical to ensure the reliability of our results.

- **Feature Selection:** EDA helps us identify which features (variables) are most relevant for your analysis or predictive modeling. We can prioritize the features that have the most impact on the outcome of interest.

- **Pattern Recognition:** EDA helps us detect patterns and trends in the data. This can include identifying seasonality, trends over time, or correlations between variables. Recognizing these patterns can guide further analysis or decision-making.

- **Outlier Detection:** EDA helps us find outliers or anomalies in the data. These outliers can be errors or may represent important observations that require special consideration in our analysis.

- **Assumption Testing:** EDA allows us to check the assumptions of statistical models. For example, we can assess whether your data meets the assumptions of linear regression or normal distribution assumptions for certain statistical tests.

- **Visualization:** EDA often involves creating visualizations such as histograms, scatter plots, and box plots. Visualizations make it easier to communicate findings and insights to stakeholders who may not have expertise in data analysis.

- **Hypothesis Generation:** During EDA, we can formulate hypotheses about relationships or patterns in the data. These hypotheses can guide further analysis and experimentation.

- **Data Preprocessing:** EDA informs data preprocessing steps such as data imputation, scaling, and encoding categorical variables. Understanding the data helps you make informed decisions about how to prepare it for analysis.

- **Risk Assessment:** EDA can reveal potential risks or limitations in our data that may affect the validity of our analysis or model. Identifying these risks early allows for appropriate mitigation strategies.

EDA serves as the foundation for any data analysis or modeling task. It helps us uncover insights, assess data quality, and make informed decisions about how to proceed with our analysis, leading to more accurate and reliable results.

## 3. Most Popular Genres:

Let's obtain all the genres:

```python
def convert_to_list(genres_array):
    return [element.strip("'") for element in genres_array]

# Apply the function to the column genres and replace the column
df_F12['genres'] = df_F12['genres'].apply(convert_to_list)
df_F12
```

```python
unique_genres = set()
for genres_list in df_F12['genres']:
    unique_genres.update(genres_list)

# Let's have the unique list of genres:
unique_genres_list = list(unique_genres)
unique_genres_list
```

Now, let's obtain the the play_time hours for each genre:

```python
sum_playtime_per_genre = []
for genre in  unique_genres_list:
  # Let's suppose that we are going to look for the genre 'norm_genre' as an input from the user.
  genre_to_find = genre

  # Create a mask to select the rows that contain the desired genre and handle NaN values
  mask = df_F12['genres'].apply(lambda x: genre_to_find in x)

  # We filtrate the DataFrame with the mask
  df_f1_by_genre = df_F12[mask]

  sum_playtime_per_genre.append(sum(df_f1_by_genre['sum_playtime_forever']))
```

```python
# We create a dataframe from the lists
df_sum_playtime_genre = pd.DataFrame({'genre': unique_genres_list, 'sum_playtime_genre': sum_playtime_per_genre})
df_sum_playtime_genre
```

```python
# Sort the DataFrame by the 'sum_playtime_genre' column in descending order
df_sum_playtime_genre_sorted = df_sum_playtime_genre.sort_values(by='sum_playtime_genre', ascending=True)

# Create a vertical histogram with purple bars and organize it from highest to lowest
plt.figure(figsize=(10, 6))
plt.barh(df_sum_playtime_genre_sorted['genre'], df_sum_playtime_genre_sorted['sum_playtime_genre'], color='purple')
plt.xlabel('Sum Playtime per Genre')
plt.ylabel('Genre')
plt.title('Vertical Histogram: Sum Playtime per Genre')
plt.tight_layout()
plt.show()
```

We can see that the most popular genre is 'Action' but now, let's contrast this information with the number of video games that have each genre:

```python
# Import Organized Dataset of steam_games with year extracted from release date
df_steam_games_WY = pd.read_parquet('/content/drive/MyDrive/Colab Notebooks/df_steam_games_WY.parquet')
df_steam_games_WY
```

```python
df_steam_games_WY.dropna(subset='genres', inplace=True)
```

```python
df_steam_games_WY.info()
```

```python
unique_genres = set()
for genres_list in df_steam_games_WY['genres']:
    # If the lists are in string format, we split it into individual elements
    if isinstance(genres_list, str):
        genres_list = genres_list.strip('[]').replace("'", "").split(', ')
    unique_genres.update(genres_list)

# Convierte el conjunto de gÃ©neros Ãºnicos de nuevo a una lista
unique_genres_list = list(unique_genres)
unique_genres_list
```

```python
count_item_per_genre = []
for genre in  unique_genres_list:
  # Let's suppose that we are going to look for the genre 'norm_genre' as an input from the user.
  genre_to_find = genre

  # Create a mask to select the rows that contain the desired genre and handle NaN values
  mask = df_steam_games_WY['genres'].apply(lambda x: genre_to_find in x)

  # We filtrate the DataFrame with the mask
  df_f1_by_genre = df_steam_games_WY[mask]
  count_item_per_genre.append(len(df_f1_by_genre['id']))

count_item_per_genre
```

```python
# Crea un DataFrame a partir de las listas
df_count_item_per_genre = pd.DataFrame({'genre': unique_genres_list, 'items_per_genre': count_item_per_genre})
df_count_item_per_genre
```

```python
# Sort the DataFrame by the 'sum_playtime_genre' column in descending order
df_count_item_per_genre_sorted = df_count_item_per_genre.sort_values(by='items_per_genre', ascending=True)

# Create a vertical histogram with purple bars and organize it from highest to lowest
plt.figure(figsize=(10, 6))
plt.barh(df_count_item_per_genre_sorted['genre'], df_count_item_per_genre_sorted['items_per_genre'], color='pink')
plt.xlabel('Items per Genre')
plt.ylabel('Genre')
plt.title('Vertical Histogram: Items per Genre')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create subplots for the two histograms side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot the sum of playtime per genre on the first subplot
ax1.barh(df_sum_playtime_genre_sorted['genre'], df_sum_playtime_genre_sorted['sum_playtime_genre'], color='purple', label='Sum Playtime')
ax1.set_xlabel('Sum Playtime')
ax1.set_ylabel('Genre')
ax1.set_title('Sum Playtime per Genre')
ax1.legend()

# Plot the number of items per genre on the second subplot
ax2.barh(df_count_item_per_genre_sorted['genre'], df_count_item_per_genre_sorted['items_per_genre'], color='pink', label='Item Count')
ax2.set_xlabel('Item Count')
ax2.set_ylabel('Genre')
ax2.set_title('Item Count per Genre')
ax2.legend()

plt.tight_layout()
plt.show()
```

It is better to do a comparison if we build the histograms in base 100%:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Calculate the total sum of playtime and total item count
total_playtime = df_sum_playtime_genre_sorted['sum_playtime_genre'].sum()
total_items = df_count_item_per_genre_sorted['items_per_genre'].sum()

# Calculate the percentages
playtime_percentages = (df_sum_playtime_genre_sorted['sum_playtime_genre'] / total_playtime) * 100
item_percentages = (df_count_item_per_genre_sorted['items_per_genre'] / total_items) * 100

# Create subplots for the two histograms side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot the sum of playtime per genre in percentage on the first subplot
ax1.barh(df_sum_playtime_genre_sorted['genre'], playtime_percentages, color='purple', label='Sum Playtime')
ax1.set_xlabel('Percentage of Total Playtime')
ax1.set_ylabel('Genre')
ax1.set_title('Percentage of Total Playtime per Genre')
ax1.legend()

# Plot the number of items per genre in percentage on the second subplot
ax2.barh(df_count_item_per_genre_sorted['genre'], item_percentages, color='pink', label='Item Count')
ax2.set_xlabel('Percentage of Total Item Count')
ax2.set_ylabel('Genre')
ax2.set_title('Percentage of Total Item Count per Genre')
ax2.legend()

plt.tight_layout()
plt.show()
```

If we compare both histograms we can see that the genre action has a high playtime because is a genre that is present in almost 17% of the items. However, we can see that multiplayer games are only 2% of the items but represent almost 8% of the user's playtime. It would be a good strategy to keep launching multiplayer videogames.

## 4. Titles Word Cloud:

```python
from wordcloud import WordCloud
```

```python
# We have a DataFrame named df_steam_games_WY
titles = df_steam_games_WY['title']

# Filter out None values and convert to a list of strings
valid_titles = [title for title in titles if isinstance(title, str)]

# Concatenate the valid titles into a single string
all_titles = ' '.join(valid_titles)

# Define a custom color function for the word cloud
def pink_purple_blue_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return f'hsl({random_state.randint(260, 320)}, {random_state.randint(60, 100)}%, {random_state.randint(50, 70)}%)'

# Create the word cloud with the custom color function
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis', max_words=200, color_func=pink_purple_blue_color_func).generate(all_titles)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Game Titles')
plt.show()
```

We can see that the words that are more repetitive in the videogames titles are:

- **Soundtrack:** Suggests that many of the games may offer a soundtrack or music-related content.
- **Pack:** Implies that there might be downloadable content (DLC) packs associated with these games.
- **Fantasy:** Could indicate that a significant number of titles are within the fantasy genre.
- **Ground:** Might suggest a common setting or theme related to the ground or terrain.
- **Rocksmith:** Could refer to a specific game title or franchise.
- **DLC:** Signifies that downloadable content is a common feature among these games.

This excercise helped us to understand the patterns and recurring elements within video game titles and help us to do a first categorization of waht type of genres are most popular. As we could see before, we have a lot of words that implies that the game are related with action or adventure.
