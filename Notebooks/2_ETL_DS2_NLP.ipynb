### 1. Data Load:

```python
import pandas as pd
import json
import ast

# Let's open the file "user_reviews" in .json format and save it into a dataframe called "df_user_reviews"
data = []
with open('Dataset/user_reviews.json', 'r', encoding='utf-8') as file:
    for line in file.readlines():
        data.append(ast.literal_eval(line))

df_user_reviews = pd.DataFrame(data)
```

### 3. Data Transformation:

We need to separate the list in the column "reviews" for each user_id. What we are going to do is to expand each user_id row with each element of the list. For this purpose, we use the function "explode".

```python
df_user_reviews_exploded = df_user_reviews.explode('reviews')
df_user_reviews_exploded
```

we can check that the user_id rows were expanded according to the number of list elements. For instance, the user_id 76561197970982479 was expanded three times.

Now we need to convert the keys of each ditionary into other columns in the dataframe. For this purpose, we are going to use the function json_normalize for Pandas as follows:

```python
df_user_reviews_normalized = pd.json_normalize(df_user_reviews_exploded['reviews'])
df_user_reviews_normalized
```

Now, with the "reviews" column transformed into a Dataframe, we need to concatenate it with "df_user_reviews_exploded" using the fuunction concat from Pandas.

```python
# We need to get rid of the indexes in both dataframes first:

df_user_reviews_exploded.reset_index(drop=True, inplace=True)
df_user_reviews_normalized.reset_index(drop=True, inplace=True)
```

```python
#Concat both dataframes:
df_user_reviews_expanded = pd.concat([df_user_reviews_exploded.drop('reviews',axis=1),df_user_reviews_normalized], axis=1)
df_user_reviews_expanded
```

### 4. Sentiment Analysis:

To do a sentiment analysis of a text we can use the library [vaderSentiment](https://github.com/cjhutto/vaderSentiment). 

The VADER (Valence Aware Dictionary and sEntiment Reasoner) library is a tool for analyzing and quantifying the sentiment or emotional tone of text data. It is specifically designed to work with text data that is often found in social media, online reviews, and informal communication, where traditional sentiment analysis methods may struggle due to the use of slang, abbreviations, emojis, and other language nuances.

It has a function that is called "sentiment.polarity_scores". It can be programmed as follows:

```python
#Import the library
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
sentiment = SentimentIntensityAnalyzer()
text_1 = "The book was a perfect balance between wrtiting style and plot."
text_2 =  "The pizza tastes terrible."
sent_1 = sentiment.polarity_scores(text_1)
sent_2 = sentiment.polarity_scores(text_2)
print("Sentiment of text 1:", sent_1)
print("Sentiment of text 2:", sent_2)
```

According to the [documentation](https://github.com/cjhutto/vaderSentiment), this is the interpretation of the scoring:

* The ``compound`` score is computed by summing the valence scores of each word in the lexicon, adjusted according to the rules, and then normalized to be between -1 (most extreme negative) and +1 (most extreme positive). This is the most useful metric if you want a single unidimensional measure of sentiment for a given sentence. Calling it a 'normalized, weighted composite score' is accurate. 
 
  It is also useful for researchers who would like to set standardized thresholds for classifying sentences as either positive, neutral, or negative.  
  Typical threshold values (used in the literature cited on this page) are:

  - **positive sentiment**: ``compound`` score >=  0.05
  - **neutral  sentiment**: (``compound`` score > -0.05) and (``compound`` score < 0.05)
  - **negative sentiment**: ``compound`` score <= -0.05
 
  **NOTE:** The ``compound`` score is the one most commonly used for sentiment analysis by most researchers, including the authors.

* The ``pos``, ``neu``, and ``neg`` scores are *ratios for proportions of text that fall in each category* (so these should all add up to be 1... or close to it with float operation).  These are the most useful metrics if you want to analyze the context & presentation of how sentiment is conveyed or embedded in rhetoric for a given sentence. For example, different writing styles may embed strongly positive or negative sentiment within varying proportions of neutral text -- i.e., some writing styles may reflect a penchant for strongly flavored rhetoric, whereas other styles may use a great deal of neutral text while still conveying a similar overall (compound) sentiment. As another example: researchers analyzing information presentation in journalistic or editorical news might desire to establish whether the proportions of text (associated with a topic or named entity, for example) are balanced with similar amounts of positively and negatively framed text versus being "biased" towards one polarity or the other for the topic/entity.

  * **IMPORTANTLY:** these proportions represent the "raw categorization" of each lexical item (e.g., words, emoticons/emojis, or initialisms) into positve, negative, or neutral classes; they **do not** account for the VADER rule-based enhancements such as word-order sensitivity for sentiment-laden multi-word phrases, degree modifiers, word-shape amplifiers, punctuation amplifiers, negation polarity switches, or contrastive conjunction sensitivity.


Now that we have understood how we can leverage this library for our analysis, let's try to use it with our first review in the dataframe:

```python
# Let's do the exercise with the first row of reviews
testSA = df_user_reviews_expanded.iloc[0,8]
text = testSA
text
```

```python
# We obtain the sentiment scores
sent_1 = sentiment.polarity_scores(text)
print("Sentiment of text:", sent_1)
```

```python
# We are going to use the compound score to classify the review in the categories 0,1 and 2. We can found this value
# in the 'compound' key of the dictionary
sent_1['compound']
```

Now it's time to do the same for all the dataframe:

```python
sentiment_score = []
for text in df_user_reviews_expanded['review']:
    if isinstance(text, str):
        sent = sentiment.polarity_scores(text)
        sentiment_score.append(sent['compound'])
    else:
        # Handle cases where 'review' is not a string (e.g., NaN values)
        sentiment_score.append(0.0)  


sentiment_score[:5]
```

Let's create a copy of our original expanded data set in order to add the column called "sentiment_score":

```python
df_user_reviews_expanded1 = df_user_reviews_expanded.copy()
df_user_reviews_expanded1 
```

```python
df_user_reviews_expanded1['sentiment_score'] = sentiment_score
```

```python
df_user_reviews_expanded1
```

We can use the "sentiment_score" to do the final classification into 0,1,2 considering the following rules as the documentation suggests:

  - **positive sentiment**: ``compound`` score >=  0.05
  - **neutral  sentiment**: (``compound`` score > -0.05) and (``compound`` score < 0.05)
  - **negative sentiment**: ``compound`` score <= -0.05

```python
sentiment_analysis = []
for score in df_user_reviews_expanded1['sentiment_score']:
    if score >= 0.05:
        sent = 2
    elif score > -0.05 and score < 0.05:
        sent = 1
    else:
        sent = 0
    sentiment_analysis.append(sent)
```

```python
df_user_reviews_expanded1['sentiment_analysis'] = sentiment_analysis
df_user_reviews_expanded1
```

```python
columns_to_eliminate = ['review','sentiment_score']
df_user_reviews_final = df_user_reviews_expanded1.drop(columns=columns_to_eliminate)
df_user_reviews_final
```
