**Step 1:**

df_steam_games has a column called `release_date`. We need to extract the year of this release date and place it in a new column called `release_year`. We ar going to do something similar as in the step 1.

```python
df_steam_games
```

As in the step 1, it is a good practice to check the NaN in the column 'release_date'.

```python
mask = df_steam_games['release_date'].isna()
df_steam_games[mask]
```

In this case, we can see that though the 'release_date' is in NaN, the entries have other information that can be useful.

Now, let's review the format of the dates in the column 'release_date', to check how can we extract the year.

```python
df_steam_games['release_date'][:60]
```

In the sample of the first 60 rows we can see that the normal format is YYYY-MM-DD. However, there are some NaN and strings that need to be considered. Also, there are entries with only the year. So we need to consider also this.

```python
import re

# We create a function to extract the year considering the data that we have in the sample above.
def extract_year(date_str):
    if pd.notna(date_str):  #If the value is not NaN
        year_match = re.search(r'\b\d{4}\b', date_str) # We loo for the year following the format '\b\d{4}\b' using the function re.search()
        if year_match:
            return year_match.group()
    return np.NaN

# we apply the function to the column release_date
release_year = df_steam_games['release_date'].apply(extract_year)

# Mostrar el DataFrame resultante
release_year
```

Now we concatenate the column release_year with the dataframe steam_games:

```python
df_steam_games_WY = pd.concat([df_steam_games, pd.DataFrame({'release_year': release_year})], axis=1)
df_steam_games_WY
```

Now we need to change the name of the column 'id' by 'item_id' to have a standard.

```python
df_steam_games_WY.rename(columns={'id': 'item_id'}, inplace=True)
df_steam_games_WY
```

```python
df_steam_games_WY.info()
```

Now we need to keep minimum columns that are requiered by the Machine Learning Model in order to save resources.

Analyzing the functions, the only columns that we are going to need are the following:

- item_id: It is important for the ML model where the user needs to enter an item_id in order to receive 5 games that are similar to the one selected.
- title: It is important for the ML model where the user needs to enter an item_id in order to receive 5 games that are similar to the one selected.
- release_year: This data can be useful because there are some gamers that prefer some vintage games better than the modern ones.
- genres: This can help us as a characterization of the video game.

- tags: This can help us as a characterization of the video game.

- developer: This can help us as a characterization of the video game.

```python
selected_columns = ['item_id','title','release_year','genres','publisher','tags','specs','developer']
df_FML1 = df_steam_games_WY[selected_columns]
df_FML1
```

First, let's check if there is any item_id in NaN:

```python
mask = df_FML1['item_id'].isna()
df_FML1[mask]
```

Let's eliminate the item_id in NaN

```python
df_FML1.dropna(subset=['item_id'], inplace=True)
df_FML1
```

```python
df_FML1.reset_index(drop=True, inplace=True)
df_FML1
```

In the data provided we can check that there are some columns that can have other uses as a filtering for a better recommendation system, and not for training our model. Let's go deeper in this analysis:

1. **release_date:** We could potentially use the release_date column to filter recommendations based on the release date, so that we recommend newer or older games based on user preferences. For example, we might give more weight to games released in a similar time frame.

2. **publisher and developer:** We could potentially use the publisher and developer column to as filters in  order provide recommendations based on user preferences for specific publishers or developers. These columns tend to be very specific and it will not allow our model to generalize.

We need to remember that including additional columns in our recommendation system can make it more sophisticated and potentially improve the quality of recommendations. However, it also increases the complexity of our feature vectors and can make the vectorization process more challenging.

Now, we have these columns that are general properties that help us to describe better a videogame
1. gernes.
2. tags.
3. specs.

However, there are a lot of information inside of those columns, so the time and resources used for building the vectors, are going to be enormous. Taking this in consideration, we can check that the column 'tags' sometimes has information from the genres and the specs (Most characteristics to describe each item). So, to have a good model with the less of the resources consumption, we consider that using only the column 'tags' may be a good option.  

```python
df_FML1_for_Model = df_FML1[['item_id','title','tags']]
df_FML1_for_Model
```

Now, we need to eliminate duplicates in order to have a better model:

```python
df_FML1_for_Model.drop_duplicates(inplace=True)
df_FML1_for_Model
```

We see that there's only one duplicate. Now, we are going to eliminate NaNs, considering that will not give value for our Recommendation System.

```python
df_FML1_for_Model.dropna(inplace=True)
df_FML1_for_Model
```

Now, we have 29,922 entries without any NaN or duplicates. Finally, we need to check the type or our variables in the dataframe:
